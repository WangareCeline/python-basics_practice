## Customer class

```python
class MyCustomer: # declares a new class called mycustomer
    def __init__(self, name, email, spend=0): # called when creating a new object(sonstructor/ initializer)
        self.name = name #instance attributes 
        self.email = email
        self.spend = spend

    def display_info(self): # returns a string that displays the information
        return f"{self.name} -Total spent {self.spend}"

    def add_purchase(self, amount): # updates the amount spent
        if amount > 0:
            self.spend = self.spend + amount
        else:
            return "Insufficient Balance"
```

```python
customer1 = MyCustomer('Anne', 'anne@gmail.com')
customer1.add_purchase(5000)
customer1.display_info()
```

## Product class

```python
class Product:
    def __init__(self, sku, name, category, price, quantity):
        self.sku = sku
        self.name = name
        self.category = category
        self.price = price
        self.quantity = quantity

    def total_value(self):
        return self.price * self.quantity
    
    def restock(self, restock_amount):
        self.quantity = self.quantity + restock_amount
        return self.quantity
```

```python
product1 = Product('w102', 'Maize', 'Food', 200, 50)
product1.total_value()
```

```python
product1.restock(4500)
product1.total_value()
```

## Student class

```python
class Student:
    def __init__(self, name, course, marks=None):
        self.name = name
        self.course = course
        self.marks = marks or []

    def average(self):
        if not self.marks: # This student has no marks yet
            return 0
        return sum(self.marks)/ len(self.marks)
    
    def grade(self):
        avg = self.average()
        if avg >= 70:
            return 'A'
        if avg >= 60:
            return 'B'
        if avg >= 50:
            return 'C'
        else:
            return 'D'
```

```python
students1 = Student('Jack', 'Machine Learning - sem1', [45,78,89,78,56,22,23])
students1.average()
students1.grade()
```

```python
student2 = Student('Janet', 'Machine Learning - sem1', [90])
student2.average()
student2.grade()
```

## Assignment
- Create a bank account class
- display balance 

```python
class BankAccount:
    def __init__(self, name, acc_number, balance, acc_type):
        self.name =name
        self.acc_number = acc_number
        self.balance = balance
        self.acc_type = acc_type

    def account_details(self):
        return f"Name: {self.name}, Account Number: {self.acc_number}, Balance: {self.balance}, Account Type: {self.acc_type}"

    def check_balance(self):
        return f"Account Number: {self.acc_number}, Balance: {self.balance}"
    
    def deposit(self, deposit_amount):
        self.balance += deposit_amount
        return self.balance
    
    def withdraw(self, withdraw_amount):
        self.balance -= withdraw_amount
        return self.balance
```

```python
client1 = BankAccount('Marcus', 12345823456123, 500, 'Savings')
client1.account_details()
```

```python
client1.check_balance()
```

```python
client1.deposit(1000)
client1.check_balance()
```

```python
client2 = BankAccount('Marcus', 12345823456123, 9000, 'Checking')
withdraw_amount = 1000
new_amount = client2.withdraw(withdraw_amount)
print(f"Withdrawn Amount: {withdraw_amount} \nNew Balance: {new_amount}")
```
