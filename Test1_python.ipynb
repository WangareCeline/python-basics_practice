## Conditional Statements (if / if-else)

```python
user_password = input("Enter your password: ")
print(user_password)
if user_password == "pass":
  print("loged in successfully")
else:
  print("Error! Wrong Password")
```

```python
# Ask the user (via input) to enter a number. If it's positive, print “Positive”, else print “Non-Positive”.
number = float(input("Enter a number: "))
if number >= 0:
  print("Positive")
else:
  print("Negative")
```

```python
#Write a program that checks if a number is even or odd.
number_to_check= 19.89
if number_to_check % 2 == 0 :
  print('Even')
else:
  print('Odd')
```

```python
# Ask the user for their age. If age ≥ 18, print “You are an adult”, else print “You are a minor”.
user_age = float(input('Enter your age as a number:  '))
if user_age >= 18:
  print("You are an adult")
else:
  print("You are a minor")
```

```python
# Given three variables x, y, z, check which one is the greatest and print it.
x, y, z = 12, 657, 897
print (max(x, y, z))
```

```python
#Ask the user to input a temperature in Celsius. If the temperature is > 30, print “Hot”, if it’s between 15 and 30 (inclusive), print “Warm”, else print “Cold”.
temp = float(input("Input a temperature in Celcius: "))
if temp > 30:
  print("Hot")
elif 30 >= temp >= 15:
  print("Inclusive")
else:
  print("Cold")
```

## Lists

```python
# Create a list fruits = ["apple", "banana", "cherry"]. Print the first and last fruit.
fruits = ["apple", "banana", "cherry"]
first_fruit = fruits[0]
last_fruit = fruits[-1]
print(first_fruit,"\n",last_fruit)
```

```python
# Add a new fruit “orange” to the list.
fruits = ["apple", "banana", "cherry"]
fruits.append("watermelon")
print(fruits)
```

```python
# Remove “banana” from the list.
fruits = ["apple", "banana", "cherry"]
fruits.remove('banana')
print(fruits)
```

```python
# Given a list of numbers [1, 2, 3, 4, 5], calculate and print their sum.
numbers = [1, 2, 3, 4, 5]
numbers_total = sum(numbers)
print(numbers_total)
```

```python
# Given a list of strings ["hello", "world"], join them into a single string with a space: “hello world”.
greeting = ["hello", "world"]
print(greeting[0]+ " "+ greeting[1])
```
