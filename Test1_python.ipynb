```python
# Given a list of numbers [1, 2, 3, 4, 5], calculate and print their sum.
numbers = [1, 2, 3, 4, 5]
numbers_total = sum(numbers)
print(numbers_total)
```

```python
# Write a program that takes a list of numbers from the user (comma separated), converts them into a list of integers, and prints only the even numbers.
user_numbers = input("Enter your numbers: eg 5, 6, 34\n")
individual_number_str = user_numbers.split(',')

# Convert string numbers to integers, handling potential whitespace and non-numeric input
numbers = []
for num_str in individual_number_str:
    try:
        numbers.append(int(num_str.strip()))
    except ValueError:
        print(f"Warning: '{num_str}' is not a valid number and will be skipped.")

# Filter for even numbers
even_numbers = [num for num in numbers if num % 2 == 0]
print(even_numbers)
```
