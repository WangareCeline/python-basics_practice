```python
# Create two variables a and b. Assign a = 10 and b = 3.14. Print both and their types.
a= 10
b= 3.14
print(a, type(a))
print(b, type(b))
```

```python
# Define a boolean variable is_student = True. Print a sentence like “Am I a student? True”.
is_student = True
print("Am I a student? ",is_student)
```

```python
# Create a variable x = None. Print it and check its type.
x = None
print(x, type(x))
```

```python
# Swap the values of two variables m = 5 and n = 8 without using a third variable.
m = 5
n = 8
m, n= n, m
print("m =",m,"\nn =",n)
```

## Conditional Statements (if / if-else)

```python
#Ask the user (via input) to enter a number. If it's positive, print “Positive”, else print “Non-Positive”.
user_input = float(input("Enter a number:/n"))
if user_input >= 0:
    print('Positive')
else:
    print('Non-Positive')
```
