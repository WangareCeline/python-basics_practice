## loops

#### Normal way

```python
numbers = [1, 2, 3, 4]
squares = []
for number in numbers:
    square = number * number
    squares.append(square) 
squares
```

#### Comprehension method

```python
square_list = [number*number for number in numbers]
square_list
```

```python
# get squares for even numbers only
even_square_list = [number*number for number in numbers if number %2 == 0]
even_square_list
```

## Assignment
1. Create a list of even numbers
2. Convert temperatures from celcius to farenheit
3. Create welcome emails (list of emails then send messages to each of them)

```python
# Create a list of even numbers
even_numbers = []
for i in range(11): #the range is 0-10
    if i % 2 == 0:
        even_numbers.append(i)
        i+=1 
    else:
        i+=1

print(even_numbers)    
```

```python
# Convert temperatures from celcius to farenheit
temp_celcius = [32,43,37,78,23,0]
temp_farenheit = []

A = 9/5
B = 32
for temp in temp_celcius:
    temp = (temp*A) + B
    temp_farenheit.append(temp)    

print(temp_farenheit)
```

```python
# Create welcome emails (list of emails then send messages to each of them)
email = ['jane@gmail.com', 'anne@gmail.com', 'joy@gmail.com']
greeting = "I hope you are well"
for acc in email:
    print(f"Welcome {acc}. {greeting}")
```

## Dictionary

#### Normal way

```python
num = [1, 2, 3, 4, 5, 6, 8, 9]
square_dict = {}
for n in num:
    square_dict[n] = n*n
square_dict
```

#### Comprehension method

```python
square_dicttionary = {n: n*n for n in num}
square_dicttionary
```

```python
products = ['Laptop', 'Mouse', 'Keyboard']
prices = [999, 29, 78]
price_catalogue ={product:price for product, price in zip(products, prices) }
price_catalogue
```

## Assignment
1. Apply discount for expensive products
2. Write a list of student names and their scores
- create grade dictionary
- add letter grades

```python
# Apply discount for expensive products
discounted = {product:price*0.9 if price>100 else price for product, price in price_catalogue.items()}
discounted
```

```python
# Write a list of student names and their scores {create grade dictionary add letter grades}
students = {'Jane': 56,
            'Anne': 94,
            'Joy': 87
            }
graded_students = {}

for name, score in students.items():
        if score >= 80:
            grade = 'A'
        elif score >= 70:
            grade = 'B'
        elif score >= 60:
            grade = 'C'
        elif score >= 50:
            grade = 'D'
        else:
            grade = 'F'
        # Add the name, score, and grade to the new dictionary
        graded_students[name] = (score, grade)

print(graded_students)
```
